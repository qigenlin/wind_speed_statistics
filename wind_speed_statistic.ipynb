# 最大风速

```r
max_speed <- function(x){
    return(max(x, na.rm = TRUE))
}

wind_day <- function(x){
    return(sum(x >= 10.8,na.rm = TRUE))
}
```

## 国家站最大风速和大风日数统计

```r
#为了与灾情数据的年份匹配，从1984年开始统计
start_year = 1984
end_year = 2015

#建立两个存统计结果的表
national_max_wind_result <- NULL
national_wind_day_result <- NULL

national_max_wind_result <- matrix(data = NA, nrow = length(start_year:end_year), ncol = length(national_wind_station)+1)
colnames(national_max_wind_result) <- rep("na",dim(national_max_wind_result)[2])
colnames(national_max_wind_result)[1] <- "year"
colnames(national_max_wind_result)[2:dim(national_max_wind_result)[2]] <- national_wind_station
national_max_wind_result[,1] <- start_year:end_year 

national_wind_day_result <- matrix(data = NA, nrow = length(start_year:end_year), ncol = length(national_wind_station)+1)
colnames(national_wind_day_result) <- rep("na",dim(national_wind_day_result)[2])
colnames(national_wind_day_result)[1] <- "year"
colnames(national_wind_day_result)[2:dim(national_wind_day_result)[2]] <- national_wind_station
national_wind_day_result[,1] <- start_year:end_year
```

```r
head(national_max_wind_result)
```

```r
head(national_wind_day_result)
```

```r
station_data <- NULL
ts_station_data <- NULL
ts_wind_speed_result <- NULL
ts_wind_day_result <- NULL
col_num <- NULL

for (i in 1:length(national_wind_station)) {
    
    station_data <- national_wind_data[national_wind_station[i] == national_wind_data$station, c("time","max_mean_wind")]
    station_data[,1] <- as.Date(as.character(station_data[,1]),"%Y-%m-%d")
    
    ts_station_data <- xts(station_data$max_mean_wind,station_data$time)
    
    ts_wind_speed_result <- apply.yearly(ts_station_data,max_speed)
    ts_wind_day_result <- apply.yearly(ts_station_data,wind_day)
    
    ts_wind_speed_result <- window(ts_wind_speed_result,start = as.Date('1984-12-31'), end = as.Date('2015-12-31'))
    ts_wind_day_result <- window(ts_wind_day_result,start = as.Date('1984-12-31'), end = as.Date('2015-12-31'))
    
    national_max_wind_result[,i+1] <- as.data.frame(ts_wind_speed_result)[,1]
    national_wind_day_result[,i+1] <- as.data.frame(ts_wind_day_result)[,1]
    
}
```

## 最大风速拟合

```r
#建立空矩阵，保存拟合结果和分布检验结果,选择5年，10年，30年和50年4个年遇型大风进行分析
#选择"weibull", "gamma", "lnorm", "norm"四个分布做拟合

wind_fit_weibull <- NULL
wind_fit_gamma <- NULL
wind_fit_lnorm <- NULL
wind_fit_norm <- NULL
wind_fit_aic <- NULL

wind_fit_weibull <- matrix(data = NA, nrow = dim(max_wind_result)[2], ncol = 5)
colnames(wind_fit_weibull) <- rep("na",dim(wind_fit_weibull)[2])
colnames(wind_fit_weibull)[1] <- "station"
colnames(wind_fit_weibull)[2:dim(wind_fit_weibull)[2]] <- c("five","ten","thirty","fifty")
wind_fit_weibull[,1] <- colnames(max_wind_result)

wind_fit_gamma <- matrix(data = NA, nrow = dim(max_wind_result)[2], ncol = 5)
colnames(wind_fit_gamma) <- rep("na",dim(wind_fit_gamma)[2])
colnames(wind_fit_gamma)[1] <- "station"
colnames(wind_fit_gamma)[2:dim(wind_fit_gamma)[2]] <- c("five","ten","thirty","fifty")
wind_fit_gamma[,1] <- colnames(max_wind_result) 

wind_fit_lnorm <- matrix(data = NA, nrow = dim(max_wind_result)[2], ncol = 5)
colnames(wind_fit_lnorm) <- rep("na",dim(wind_fit_lnorm)[2])
colnames(wind_fit_lnorm)[1] <- "station"
colnames(wind_fit_lnorm)[2:dim(wind_fit_lnorm)[2]] <- c("five","ten","thirty","fifty")
wind_fit_lnorm[,1] <- colnames(max_wind_result)

wind_fit_norm <- matrix(data = NA, nrow = dim(max_wind_result)[2], ncol = 5)
colnames(wind_fit_norm) <- rep("na",dim(wind_fit_norm)[2])
colnames(wind_fit_norm)[1] <- "station"
colnames(wind_fit_norm)[2:dim(wind_fit_norm)[2]] <- c("five","ten","thirty","fifty")
wind_fit_norm[,1] <- colnames(max_wind_result)

wind_fit_aic <- matrix(data = NA, nrow = dim(max_wind_result)[2], ncol = 5)
colnames(wind_fit_aic) <- rep("na",dim(wind_fit_aic)[2])
colnames(wind_fit_aic)[1] <- "station"
colnames(wind_fit_aic)[2:dim(wind_fit_aic)[2]] <- c("weibull", "gamma", "lnorm", "norm")
wind_fit_aic[,1] <- colnames(max_wind_result)
```

```r
for  (i in 1:length(colnames(max_wind_result))) {
    
    fit_weibull <- fitdist(max_wind_result[,i], "weibull")
    fit_gamma <- fitdist(max_wind_result[,i], "gamma")
    fit_lnorm <- fitdist(max_wind_result[,i], "lnorm")
    fit_norm <- fitdist(max_wind_result[,i], "norm")
    
    wind_fit_weibull[i,2:5] <- unlist(quantile(fit_weibull, probs = c(0.8,0.9,0.967,0.98))$quantiles)   
    wind_fit_gamma[i,2:5] <- unlist(quantile(fit_gamma, probs = c(0.8,0.9,0.967,0.98))$quantiles)
    wind_fit_lnorm[i,2:5] <- unlist(quantile(fit_lnorm, probs = c(0.8,0.9,0.967,0.98))$quantiles)
    wind_fit_norm[i,2:5] <- unlist(quantile(fit_norm, probs = c(0.8,0.9,0.967,0.98))$quantiles)
    
    wind_fit_aic[i,2] <- fit_weibull$aic
    wind_fit_aic[i,3] <- fit_gamma$aic
    wind_fit_aic[i,4] <- fit_lnorm$aic
    wind_fit_aic[i,5] <- fit_norm$aic
    
    rm(fit_weibull,fit_gamma,fit_lnorm,fit_norm)
}
```

```r
#根据AIC值判断哪个拟合分布结果较好
optimal_distribution <- apply(wind_fit_aic[,2:5], 1, which.min)
```
